```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
#import csv
FedRates_file = "Resources/index 2.csv"
FedRates_df = pd.read_csv(FedRates_file)
FedRates_df.head()
```

```python
#import csv 2
Debt_file = "Resources/Transposed_Debt.csv"
Debt_df = pd.read_csv(Debt_file)
Debt_df.head(78)
```

```python
#import csv 3
Debt_file2 = "Resources/HHD_C_Report_2020Q3.csv"
Debt_df2 = pd.read_csv(Debt_file2)
Debt_df2.head()
```

```python
# Cleading Fed table

#Eliminate columns we do not need
FedRates_df = FedRates_df[['Year', 'Month', 'Day', 'Effective Federal Funds Rate', 'Unemployment Rate']]
FedRates_df.head()
```

```python
# cleaning debt2
Debt_df2 = Debt_df2.rename(columns={
                        'Total Debt Balance and Its Composition': 'quarter',
                        'Unnamed: 1':'mortgage', 'Unnamed: 2':'he_revolving', 'Unnamed: 3':'auto_loan', 
                         'Unnamed: 4': 'credit_card', 'Unnamed: 5': 'student_loan', 'Unnamed: 6':'other', 
                         'Unnamed: 7':'total'})
# cleaning debt2
Debt_df2 = Debt_df2[['quarter','mortgage','he_revolving', 'auto_loan', 'credit_card', 
                    'student_loan', 'other' , 'total']]

Debt_df2 =Debt_df2.iloc[3:]
#Debt_df2 = Debt_df2.set_index('quarter')
Debt_df2 = Debt_df2.dropna()
Debt_df2.head(70)
```

```python
Debt_df =Debt_df.rename(columns={
'Unnamed: 0': 'quarter', 'Mortgage':'mortgage', 
    'HE Revolving' :'he_revolving', 
    'Auto Loan': 'auto_loan',
    'Credit Card': 'credit_card',
    'Other':'other',
    'Total':'total',
    
})
#Debt_df = Debt_df.set_index('quarter')
Debt_df['mortgage'] = Debt_df['mortgage'].astype(float)
Debt_df['he_revolving'] = Debt_df['he_revolving'].astype(float)
Debt_df.head(17)
```

```python
#student loan debt does not appear to be in other before 03

frames = [Debt_df, Debt_df2]
result = pd.concat(frames)

result['quarter'] = result['quarter'].apply(lambda x: x.replace(':','_'))
result.head(89)
```

```python
#Converting interest rate to a 2 dimensional list we can iterate over
subsetFedData = FedRates_df[FedRates_df.Year > 1998]

subsetFedData.groupby('Year').count()
```

```python
#Get all rows that have a year greater than 1998 (This is our desired data range).
interestRateList = subsetFedData.values.tolist()

print(interestRateList)
```

```python
quarterizedList = []
interestRateSum = 0.0
averagedRate = 0.0
newQuarter = ""
newYear = ""

for dataRow in interestRateList:    
    if dataRow[1] == 1:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 2:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 3:
        #This should totally be in a function, but it's Saturday and I don't feel like doing that
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q1"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 4:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 5:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 6:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3] 
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q2"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 7:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 8:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 9:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q3"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 10:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 11:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 12:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q4"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""          
        
```

```python
finalFedFrame = pd.DataFrame(quarterizedList, columns=['quarter', 'rate'])

finalFedFrame.head()
```
