```python
#Converting interest rate to a 2 dimensional list we can iterate over
subsetFedData = FedRates_df[FedRates_df.Year > 1998]

subsetFedData.groupby('Year').count()
```

```python
#Get all rows that have a year greater than 1998 (This is our desired data range).
interestRateList = subsetFedData.values.tolist()

print(interestRateList)
```

```python
quarterizedList = []
interestRateSum = 0.0
averagedRate = 0.0
newQuarter = ""
newYear = ""

for dataRow in interestRateList:    
    if dataRow[1] == 1:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 2:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 3:
        #This should totally be in a function, but it's Saturday and I don't feel like doing that
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q1"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 4:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 5:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 6:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3] 
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q2"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 7:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 8:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 9:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q3"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""
    if dataRow[1] == 10:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]            
    if dataRow[1] == 11:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
    if dataRow[1] == 12:
        if not np.isnan(dataRow[3]):
            interestRateSum += dataRow[3]
        else:
            continue
            
        averagedRate = round((interestRateSum/3), 2)
        newYear = str(int(dataRow[0]))
        newQuarter = newYear[-2:] + "_Q4"  
        
        newRow = [newQuarter, averagedRate]
        
        quarterizedList.append(newRow)
        
        interestRateSum = 0.0
        averagedRate = 0.0
        newQuarter = ""
        newYear = ""          
        
```
