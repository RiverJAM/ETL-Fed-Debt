```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv
FedRates_file = "Resources/index 2.csv"
FedRates_df = pd.read_csv(FedRates_file)
FedRates_df.head()
```

```python
#import csv 2
Debt_file = "Resources/Transposed_Debt.csv"
Debt_df = pd.read_csv(Debt_file)
Debt_df.head(78)
```

```python
#import csv 3
Debt_file2 = "Resources/HHD_C_Report_2020Q3.csv"
Debt_df2 = pd.read_csv(Debt_file2)
Debt_df2.head()
```

```python
# Cleading Fed table

#Eliminate columns we do not need
FedRates_df = FedRates_df[['Year', 'Month', 'Day', 'Effective Federal Funds Rate', 'Unemployment Rate']]
FedRates_df.head()
```

```python
# cleaning debt2
Debt_df2 = Debt_df2.rename(columns={
                        'Total Debt Balance and Its Composition': 'quarter',
                        'Unnamed: 1':'mortgage', 'Unnamed: 2':'he_revolving', 'Unnamed: 3':'auto_loan', 
                         'Unnamed: 4': 'credit_card', 'Unnamed: 5': 'student_loan', 'Unnamed: 6':'other', 
                         'Unnamed: 7':'total'})
# cleaning debt2
Debt_df2 = Debt_df2[['quarter','mortgage','he_revolving', 'auto_loan', 'credit_card', 
                    'student_loan', 'other' , 'total']]

Debt_df2 =Debt_df2.iloc[3:]
#Debt_df2 = Debt_df2.set_index('quarter')
Debt_df2 = Debt_df2.dropna()
Debt_df2.head(70)
```

```python
Debt_df =Debt_df.rename(columns={
'Unnamed: 0': 'quarter', 'Mortgage':'mortgage', 
    'HE Revolving' :'he_revolving', 
    'Auto Loan': 'auto_loan',
    'Credit Card': 'credit_card',
    'Other':'other',
    'Total':'total',
    
})
#Debt_df = Debt_df.set_index('quarter')
Debt_df.head(17)
```

```python
#student loan debt does not appear to be in other before 03

frames = [Debt_df, Debt_df2]
result = pd.concat(frames)

result =result.apply(lambda x: x.str.replace(':','_'))
result.head(89)
```
